```python
num_list = list(range(1,409))
url_list = []

for each_number in num_list:
    each_url = 'https://search.hospitalpriceindex.com/hpi2/hospital/Los%20Angeles%20Community%20Hospital/7919?page=' + str(each_number)
    url_list.append(each_url)
    
len(url_list)
```

```python
url_list
```

```python
# list of URLS

# make a dataframe list

dataframes=[]

for each_url in url_list:
    driver.get(each_url)
    time.sleep(1)
    new_df = pd.read_html(driver.page_source)[0]
    dataframes.append(new_df)
len(dataframes)    
#dataframes[0]
    
```

```python
df = pd.concat(dataframes)
len(df)
```

```python
df.to_csv("la_community_chargemaster.csv", index=False)
```
